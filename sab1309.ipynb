## Conjunto (set)
Estructura de datos con almacenamiento y operaciones similar al de los conjuntos matemáticos.
[Referencia](https://www.w3schools.com/python/python_sets.asp)

```python
A1={"rojo","verde","azul"}
A2={"verde","negro","amarillo"}
print(type(A1))
print(A1)
```

```python
A3=A1.intersection(A2)
A4=A2.intersection(A1)
print(type(A3))
print(A3)
print(A4)
```

```python
A5=A1.difference(A2) 
print(A5)
A1.isdisjoint(A2)
respuesta=A1.isdisjoint(A3)  #Es disjunto?
print(respuesta)
```

```python
Valor=A1.__contains__("verde") #pertenece
print(Valor)
```
