```python
## **MÉTODOS Y FUNCIONES PARA LISTAS**
#Un método es una función que "pertenece" a un objeto
```

```python
edades=[25,23,12,45,13,23]
#edades.sort
edades.sort() #método sort
print(edades)
```

```python
edades=[25,23,12,45,13,23]
#edades.sort
edades.sort(reverse=True) #método sort
print(edades)
edades.append(100)
print(edades)
edades.sort()
print(edades)
```

```python
nombres = ["Carlos", "Ana", "Beatriz","X","F"] # PRUEBA CON LETRAS
nombres.sort(reverse=True)
print(nombres)
```
